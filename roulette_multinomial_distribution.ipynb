```python
from math import factorial, comb, log, exp
from collections import defaultdict
```

```python
n = int(input("Enter number of events: "))
red = 18 / 37
black = 18 / 37
zero = 1 / 37

if abs(red + black + zero - 1) > 0.0001:
    print("-" * 104)
    print(f"Win + Draw + Loss must be 100: ")
    print(f"Enter correct data please: ")
    print("-" * 104)
    exit()

potential_distribution = defaultdict(float)
multinomial_coefficient = defaultdict(int)
TOTAL_EVENTS_FACTORIAL = factorial(n)
k = 3

for red_ in range(n + 1):
    for black_ in range(n + 1):
        zero_ = n - red_ - black_

        if zero_ < 0:
            continue

        coefficient = TOTAL_EVENTS_FACTORIAL / (factorial(red_) * factorial(black_) * factorial(zero_))
        log_result = (red_ * log(red)) + (black_ * log(black)) + (zero_ * log(zero)) + log(coefficient)

        result = exp(log_result)
        distribution = (red_, black_, zero_, int(coefficient))

        potential_distribution[distribution] += result
        multinomial_coefficient[coefficient] += 1
```

```python
sorted_potential_distribution = dict(sorted(potential_distribution.items(), key=lambda kvp: (-kvp[1], kvp[0])))

print("-" * 104)
print("The combinations and their probabilities are:")

for index, (key, value) in enumerate(sorted_potential_distribution.items(), start=1):
    red, black, zero, coef = key
    print(f"#{index:02_} (red: {red}, black: {black}, zero: {zero}) -> {value:.4%}  || "
          f" Mult. coef: {coef:_}  ---  1 From: {1 / value:_.1f} trials")
print()
print(f"Total unique combinations is: {comb(n + k - 1, n):_}")
```

```python
print("-" * 104)

top_5 = list(sorted_potential_distribution.items())[:5]
cumulative = 0
print(f"Top 5 Combinations: ")
for i, (key, prob) in enumerate(top_5, start=1):
    r, b, z, coef = key
    cumulative += prob
    print(f"{i}. Red: {r:2d} -- Black: {b:2d} -- Zero: {z:2d}   |  {prob:.4%}  |  Cumulative: {cumulative:.4%}"
          f"  (1 in {1 / prob:_.1f} trials)")

print("-" * 104)
```

```python
mean_red = n * (18 / 37)
mean_black = n * (18 / 37)
mean_zero = n * (1 / 37)

print(f"Expected values (mean):")
print(f"  Red wins:   {mean_red:.2f}")
print(f"  Black wins: {mean_black:.2f}")
print(f"  Zeros:      {mean_zero:.2f}")

print("-" * 104)
```

```python
sorted_multinomial_coefficient = dict(sorted(multinomial_coefficient.items(), key=lambda x: (x[0], -x[1])))

for i, (multinomial_coef, frequency) in enumerate(sorted_multinomial_coefficient.items(), start=1):
    probability = frequency / sum(sorted_multinomial_coefficient.values())
    print(f"#{i:02} Multinomial coefficient: {multinomial_coef:_} | Frequency: {frequency} "
          f"| Probability: {probability:.4%} | 1 From {1 / probability:.1f}")

print("-" * 104)
```
